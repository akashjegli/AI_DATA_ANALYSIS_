```python
# Question 1: Importing Pandas and Loading a CSV File
import pandas as pd
df = pd.read_csv('sample_data.csv')

# Question 2: Displaying the First Few Rows
df.head()

# Question 3: Basic Data Information
df.info()

# Question 4: Inspecting Column Data Types
df.dtypes

# Question 5: Selecting Columns
df['Department']

# Question 6: Slicing Rows
df[2:5]

# Question 7: Handling Missing Values
df.fillna(0, inplace=True)

# Question 8: Renaming Columns
df.rename(columns={'Salary': 'Annual_Salary'}, inplace=True)

# Question 9: Dropping Duplicates
df.drop_duplicates(inplace=True)

# Question 10: Grouping and Aggregating Data
df.groupby('Department')['Age'].mean()

# Question 11: Exporting Data to CSV
df.to_csv('cleaned_data.csv', index=False)

# Question 12: Aggregating with Multiple Functions
df.groupby('Department').agg({'Age': ['mean', 'max'], 'Annual_Salary': ['sum', 'mean']})
```
